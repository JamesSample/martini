```python
import nivapy3 as nivapy
import pandas as pd
```

## Map rivers with CDOM data

Show locations of rivers with UV absorbance data from Cathrine (see e-mail from Phil received 04.11.2019 at 10.28).

```python
df = pd.read_excel(r'../data/elveovervakingsprogrammet_rivers.xlsx')
df
```

```python
df = nivapy.spatial.utm_to_wgs84_dd(df,
                                    zone='UTM_Zone',
                                    east='UTM_E',
                                    north='UTM_N')
df
```

```python
m = nivapy.spatial.quickmap(df,
                            lon_col='lon',
                            lat_col='lat',
                            popup='River')

m
```
